```python
myfile = open("romeoandjuliet.txt")
```

```python
uniqueWords = {}
for word in myfile.read().split():
    if word not in uniqueWords:
        uniqueWords[word] = 1
    else:
        uniqueWords[word] += 1
for k, v in uniqueWords.items():
    print(k, v)
```

```python
import matplotlib.pylab as plt
import operator
import numpy as np
import math
```

```python
sortedDict = sorted(uniqueWords.items(), key=operator.itemgetter(1))
```

```python
print(sortedDict)
```

```python
x, y = zip(*sortedDict)
```

```python
print(x, y)
```

```python
plt.bar(range(len(x)), y, 0.75)
plt.xlabel("Words", fontsize = 20)
plt.xticks(range(len(x)), x)
plt.ylabel("Frequency", fontsize = 20)
plt.ylim(0, 400)
plt.rcParams['figure.figsize'] = [15, 15]
plt.show()
```

The 10 most frequent words are "to", "a", "I", "the", "of", "and","my", "is", "in", and "that"; and the 10 least frequent words are all words that occur just once in the text. This suggests that the text is most often using articles, pronouns, and prepositions. And the least frequent are proper nouns, to adjectives, to nouns.

```python
invertedX = x[::-1]
invertedY = y[::-1]
fig1, ax1 = plt.subplots()

plt.plot(range(len(x)), invertedY, 0.75)
plt.xlabel("Words", fontsize = 20)
ax1.set_xscale('log')
ax1.set_yscale('log')
plt.ylabel("Frequency", fontsize = 20)
plt.rcParams['figure.figsize'] = [15, 15]
plt.show()
```
